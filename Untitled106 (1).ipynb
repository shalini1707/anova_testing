## anova ***testing***

```python
import pandas as pd
from scipy import stats
```

```python
df=pd.read_csv("/content/Wine_Quality_Data.csv")
df
```

```python
df_anova=df[["residual_sugar","color"]]
df_anova
```

```python
value=pd.unique(df_anova.color.values)
value
```

```python
x={}
for i in value:
  x[i]=df_anova["residual_sugar"][df_anova.color==i]
```

```python
x
```

```python
## testing packing for stats
```

```python
fval,pval=stats.f_oneway(x["red"],x["white"])
```

```python
pval
```

```python
if pval<0.05:
  print("null hypothesis h0 is rejected,ha is accepted")
  print("there is no realationship between two categorical column")
else:
  print("null hypothesis h0 is accpted,ha is rejected")
  print("there is some realationship between two categorical column")
```
